What is LFP (Local Field Potential) signal? Explain briefly  (5 points)

Your answer:LFP signals, or Local Field Potentials, are a type of electrophysiological signal commonly used in neuroscience research. They represent the extracellular electrical activity recorded from a population of neurons in a specific brain region. LFP signals provide insights into both specific brain regions and overall brain activity.

The LFP signal reflects the synchronized electrical potentials generated by nearby neurons, and it is typically measured using microelectrodes placed within brain tissue or on the surface of the brain. It represents the summed electrical activity of multiple neurons near the recording electrode.

LFP signals have a lower frequency range compared to single-neuron action potentials, typically ranging from 0.1 Hz to a few hundred Hz. They often exhibit oscillatory patterns in different frequency bands, such as theta (4-8 Hz), alpha (8-12 Hz), beta (12-30 Hz), gamma (30-100 Hz), or other bands.

Researchers analyze LFP signals to study various aspects of brain function, including sensory processing, cognition, motor control, and pathological conditions such as epilepsy or Parkinson's disease. LFP signals provide insights into network-level activity, synchronization, and information flow within the brain regions under investigation.

LFP signals can be used to study both specific brain regions and overall brain activity. When recorded from specific regions, they offer information about the local processing and functional properties of neurons. Simultaneously, LFP signals also reflect the overall brain activity, revealing network dynamics and communication between different brain regions.

Examples of cognitive processes and behaviors studied using LFP signals include attention, memory, decision-making, perception, motor control, sleep, consciousness, and neuropsychiatric disorders. LFP signals provide insights into the neural correlates of these processes, shedding light on their underlying mechanisms.

In summary, LFP signals are valuable tools in neuroscience research, allowing investigation into the dynamics of neural activity within specific brain regions and the coordination of activity between different regions during various cognitive processes and behaviors.

# Pre-processing of the data

Now we want to pre process our data. Follow the following steps and explain why we do them:  (points 20)

1. Apply a bandpass filter in the range 1-300 Hz on the signal of each region.

2. Convert the signal to 120 stimulus trials using digitalByte.

3. Apply z-score normalization on each trial.

```python
def bandpass_filter(data, low_freq, high_freq):
    nyq = 0.5 * sample_rate
    l, h = low_freq / nyq, high_freq / nyq
    b, a = sp.signal.butter(5, [l, h], btype='band')
    return sp.signal.filtfilt(b, a, data)


def extract_trials(data, digital_byte, expected_trials=120):
    trials = []
    trial_onsets = []
    for i in range(1, len(digital_byte)):
        if digital_byte[i] != 0 and digital_byte[i-1] == 0:
            trial_onsets.append(i)
    if np.abs(len(trial_onsets) - expected_trials) > 1:
        print('ERROR in extracting trials!')
        return
    for onset_index in trial_onsets:
        trial_end_index = onset_index + sample_rate * 3
        trial_data = data[onset_index:trial_end_index]
        trials.append(trial_data)
    return trials


def z_score(data):
    return (data - np.mean(data)) / np.std(data)
```

```python
filtered_signals = []
trials = []
normalized_trials = []

for i in range(3):
    filtered_signal = bandpass_filter(channel_data[:, i], 1, 300)
    trial = extract_trials(filtered_signal, digital_byte)
    normalized_trial = [z_score(t) for t in trial]
    filtered_signals.append(filtered_signal)
    trials.append(trial)
    normalized_trials.append(normalized_trial)
```

Event-Related Potentials (ERPs) are electrical signals in the brain that occur in response to specific events or stimuli. They provide valuable insights into the brain's activity during sensory, cognitive, or motor processes.

Researchers obtain ERPs by analyzing the electrical signals recorded by an electroencephalogram (EEG) from multiple repetitions or trials of the same task or stimulus. By averaging these signals, researchers can isolate the brain's response to the specific event or stimulus from background noise.

ERPs are characterized by distinctive patterns of positive and negative deflections in the EEG waveform, known as components. These components, labeled with letters or abbreviations such as P1, N2, or P300, correspond to different stages of information processing.

The timing and magnitude of ERP components provide valuable insights into various cognitive processes and neural mechanisms, including perception, attention, memory, and language processing. Researchers use ERPs to investigate brain function, including sensory and cognitive processes, emotional responses, and the effects of disorders or interventions on brain activity.

ERPs are particularly useful in experimental settings where researchers can manipulate and control the timing and presentation of stimuli. This allows them to examine how the brain responds to specific events or conditions. Through the study of ERPs, researchers gain a non-invasive understanding of the timing and dynamics of neural processing, providing important information about how the brain responds to different stimuli and cognitive tasks.
ERP Components: ERPs consist of different components that represent distinct stages of information processing in the brain. These components are identified based on their timing and polarity. For instance, the P1 component occurs approximately 80-120 milliseconds after stimulus onset and is associated with early sensory processing. The N2 component, occurring around 200-300 milliseconds, is related to cognitive processes like attention and decision-making. The P300 component, appearing around 300-500 milliseconds, is often linked to higher-level cognitive functions such as working memory and attention allocation.

Experimental Manipulation: ERPs are highly valuable in experimental settings because researchers can manipulate and control the timing and characteristics of stimuli. This control enables the investigation of specific aspects of brain function. By systematically varying the properties of stimuli or tasks, researchers can examine how the brain's ERP responses change accordingly. This approach helps identify the neural processes underlying different cognitive functions and provides insights into the timing and dynamics of information processing.

Non-Invasive Measurement: ERPs are measured using non-invasive techniques, primarily electroencephalography (EEG). EEG involves placing electrodes on the scalp to record the brain's electrical activity. This method allows for precise measurement of ERP components with high temporal resolution, capturing millisecond-level changes in brain activity. EEG is widely used in research and clinical settings due to its accessibility, affordability, and ability to provide real-time information about brain function.

Clinical Applications: ERPs have been extensively studied in various clinical populations to investigate abnormalities in brain functioning. For example, researchers have examined ERPs in individuals with neurological disorders like epilepsy, Alzheimer's disease, and attention-deficit/hyperactivity disorder (ADHD). By comparing ERPs between clinical groups and healthy individuals, researchers can identify specific ERP abnormalities associated with these conditions, aiding in diagnosis and treatment planning.

Cognitive and Brain Research: ERPs have played a significant role in advancing our understanding of cognitive processes and the underlying neural mechanisms. They provide insights into perception, attention, memory, language processing, emotion, and other mental functions. By examining ERP responses in different experimental paradigms, researchers can investigate how these processes are modulated by various factors, including task demands, stimulus characteristics, and individual differences.

Phase-phase coupling is a fundamental aspect of functional connectivity (FC) research, which examines the temporal correlations between neurophysiological events in different regions of the brain. It provides insights into how these regions communicate and synchronize their activities. Two commonly used measures to assess phase-phase coupling are Phase Locking Value (PLV) and Phase Lag Index (PLI).

PLV quantifies the degree of phase synchronization or locking between two signals across multiple time points. It measures the consistency of phase relationships, ranging from 0 to 1. A value of 0 indicates no phase coupling, while a value of 1 indicates perfect synchronization. PLV is often employed to assess the strength of phase coupling between different frequency bands or brain regions.

In contrast, PLI focuses on the consistency of phase lag rather than the exact phase relationship. It measures the asymmetry in the distribution of phase differences between two signals. PLI ranges from 0 to 1, with 0 indicating no consistent phase lag and 1 indicating a consistent phase lag. PLI is particularly useful when studying phase relationships in the presence of volume conduction, which introduces spurious correlations due to the spread of electrical activity through conductive mediums like the skull.

Both PLV and PLI are valuable measures for studying phase-phase coupling. However, they have distinct characteristics. PLV captures both zero-lag and non-zero-lag phase relationships, providing a global measure of phase coupling strength. On the other hand, PLI primarily detects non-zero-lag phase relationships and is less influenced by common sources or volume conduction biases.

The article titled "Phase lag index: Assessment of functional connectivity from multi-channel EEG and MEG with diminished bias from common sources" explores the use of PLI as a measure of functional connectivity. It addresses the issue of common sources and the potential biases they introduce in FC analysis. The authors propose PLI as a method that reduces the impact of common sources, enabling a more accurate assessment of functional connectivity.

In summary, PLV and PLI are widely used measures for examining phase-phase coupling and functional connectivity in the brain. PLV captures the synchronization of phase angles, while PLI focuses on the consistency of phase lag. Both measures provide valuable insights into brain dynamics and can be employed depending on the research question and potential biases in the data.

```python
wPLI = {}
for i in range(3):
    for j in range(i + 1, 3):
        total_num, total_denom = 0, 0
        for epoch1 in normalized_trials[i]:
            for epoch2 in normalized_trials[j]:
                n, d = WPLI([epoch1, epoch2])
                total_num += n
                total_denom += d
        wPLI[(i, j)] = np.abs(total_num) / total_denom
```

wPLI (weighted Phase Lag Index) and PLI (Phase Lag Index) are measures used in neuroimaging research to assess the functional connectivity between different brain regions based on their phase synchronization. These measures provide valuable insights into the strength and directionality of phase interactions within the brain.

Both wPLI and PLI are derived from analyzing phase differences in the brain's oscillatory activity, typically measured using techniques such as electroencephalography (EEG) or magnetoencephalography (MEG).

The main distinction between wPLI and PLI lies in how they handle the distribution of phase differences. PLI considers only the sign (positive or negative) of the phase differences, focusing on the asymmetry of phase synchrony. It is particularly useful for identifying phase-leading or phase-lagging relationships between brain regions.

In contrast, wPLI takes into account both the sign and magnitude of the phase differences. It assigns weights to the phase differences based on their magnitude, capturing the strength of phase synchronization. By considering magnitude, wPLI provides a more comprehensive measure of connectivity, incorporating information about both the presence and strength of phase synchrony between brain regions.

In summary, PLI emphasizes the asymmetry of phase synchrony, while wPLI provides a weighted estimate of functional connectivity that considers both the sign and magnitude of phase differences. These measures are valuable tools for studying brain connectivity and can be applied to investigate various cognitive and clinical phenomena.
